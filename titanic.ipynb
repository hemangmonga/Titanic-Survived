```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv("train.csv")
```

```python
df_train
```

```python
df_train = df_train.fillna(value = {'Cabin': 0, 'Embarked':'S'})

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df_train['Age'] = imputer.fit_transform(df_train[['Age']]).round()    #imputing missing values of age column

df_train.dropna(subset = ["Age"], inplace=True)


df_train = df_train.drop(columns = ['Name', 'Ticket','Cabin','Fare'])
df_train = df_train.set_index('PassengerId')
df_train
```

```python
df_train.to_csv('train_data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_train.iloc[:,-5] = le.fit_transform(df_train.iloc[:,-5])
df_train.iloc[:,-1] = le.fit_transform(df_train.iloc[:,-1])
```

```python
X_train = df_train.copy()
X_train = X_train.drop(columns = ['Survived'])
y_train = df_train['Survived']
```

```python
df_test = pd.read_csv('test.csv')
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df_test['Age'] = imputer.fit_transform(df_test[['Age']]).round()
df_test['Fare'] = imputer.fit_transform(df_test[['Fare']])
df_test =  df_test.fillna(value = {'Cabin': 0})
df_test = df_test.drop(columns = ['Name', 'Ticket','Cabin','Fare'])

df_test = df_test.set_index("PassengerId")
df_test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_test.iloc[:,1] = le.fit_transform(df_test.iloc[:,1])
df_test.iloc[:,-1] = le.fit_transform(df_test.iloc[:,-1])
```

```python
df_test.dtypes
```

```python
X_test = df_test.copy()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
def models(X_train, y_train):

    from sklearn.linear_model import LogisticRegression
    logit = LogisticRegression(random_state = 0)#Model instance
    logit.fit(X_train, y_train) #Train the model
   

    from sklearn.tree import DecisionTreeClassifier
    dtree = DecisionTreeClassifier(max_depth=28, criterion = "entropy", min_samples_split=10) #model instance
    dtree.fit(X_train, y_train) #train model

    
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    knn.fit(X_train, y_train)
    
    from sklearn.naive_bayes import GaussianNB #Import required algorithm
    NB=GaussianNB() #model instance
    NB.fit(X_train, y_train) #Train model
    
    from sklearn.ensemble import RandomForestClassifier
    rm = RandomForestClassifier(n_estimators = 10, max_depth=25, criterion = "gini", min_samples_split=10)
    rm.fit(X_train, y_train)
    
    from xgboost import XGBClassifier
    xg=XGBClassifier(learning_rate = 0.05, n_estimators=300, max_depth=5) #model instance
    xg.fit(X_train, y_train) #Train model
    
    from sklearn.svm import SVC
    svc_lin = SVC(kernel = 'linear', random_state = 0)
    svc_lin.fit(X_train, y_train)
    
    from sklearn.svm import SVC
    svc_rbf = SVC(kernel = 'rbf', random_state = 0)
    svc_rbf.fit(X_train, y_train)
    
    print("[0]Logistic Regression Accuracy:", logit.score(X_train,y_train))
    print("[1]Decision Tree Accuracy:", dtree.score(X_train,y_train))
    print("[2]KNeighbors Accuracy:", knn.score(X_train,y_train))
    print("[3]Gaussian NB Accuracy:", NB.score(X_train,y_train))
    print("[4]XG Boost Accuracy:", xg.score(X_train,y_train))
    print("[5]SVC linear Accuracy:", svc_lin.score(X_train,y_train))
    print("[6]SVC rbf Accuracy:", svc_rbf.score(X_train,y_train))
    
    
    
    return logit, dtree, knn, NB, rm, xg, svc_lin, svc_rbf
    
```

```python
model = models(X_train, y_train)
```

```python
df_test = df_test.reset_index()
```

```python
final_result = df_test['PassengerId']
pred = model[1].predict(X_test)
pred = pd.DataFrame({'PassengerId':final_result, 'Survived':pred})
```
